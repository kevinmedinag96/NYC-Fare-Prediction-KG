The objective of this notebook is to conduct an exploratory data anlysis on the data included in each parque file from the tlc nyc taxi trip webpage

```python
import requests
import pandas as pd
```

```python
nyc_2024_01_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2024-01.parquet"
columns = ["vendor_id",'tpep_pickup_datetime', 'tpep_dropoff_datetime',
       'passenger_count', 'trip_distance',"ratecode_id","store_and_fwd_flag","pu_location_id","do_location_id","payment_type",
       'fare_amount', 'extra',
       'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge',
       'total_amount', 'congestion_surcharge',"airport_fee"]
df = pd.read_parquet(nyc_2024_01_url)
df.rename({
    col_prev : columns[i] for i,col_prev in enumerate(df.columns)
},inplace=True,axis=1)
```
