```python
import json
import pandas as pd
import numpy as np
import os
import pathlib
import re

#-------------- AZURE MODULES
from azure.cosmos import CosmosClient
```

```python
curr_path = pathlib.Path.cwd()
data_path = str(curr_path.parent.parent.parent.parent) + "/data_scientist_challenge_sp/datasets/dataset-test-general.json" 
```

```python
with open(data_path, 'r') as f:
    data = json.load(f)


df = pd.DataFrame.from_dict(data,orient="columns")
df.drop("DOLocationID", axis=1,inplace=True)
df.reset_index(inplace=True,drop=True)


#include ID in database...

id_series = pd.Series(np.arange(1,df.shape[0] + 1,dtype=int),name="ItemID")

df = pd.concat([id_series,df],axis=1)
```

# Connect to Azure Cosmos DB for NOSQL

```python
os.environ["COSMOSDB_ACCNT_URI"]
URL = os.environ['COSMOSDB_ACCNT_URI']
KEY = os.environ['COSMOSDB_ACCNT_key']
client = CosmosClient(URL, credential=KEY)

azure_cosmos_db = client.get_database_client("db-1")
container = azure_cosmos_db.get_container_client("NYCTestSetSP")
```

```python
for i in range(df.shape[0]):
    row = df.iloc[i,:].to_dict()
    row["id"] = str(i + 1)
    container.create_item(row)
```

```python
res = []
mi_string = list("abcd")
n = len(mi_string)
def find_anagram(arr_string,l,r):
    global res
    if r == n-1:
        return ["".join(arr_string)]

    for r in range(l,n):
        #switch 
        arr_string[l],arr_string[r] = arr_string[r],arr_string[l]
        res = res + find_anagram(arr_string,l+1,r)
    return res

#l 3 2 1
#abcd , abdc , acbd, acdb, adbc, adcb, 

#permutations of bcd -> bcd, bdc, cbd, cdb, dbc, dcb
#permutations of abcd -> abcd, abdc, acbd, acdb, adbc, adcb
```
